```python
import pandas as pd
dataset = pd.read_csv("/content/vgsales.csv")

dataset.head(30)
```

```python
#video_games = dataset.groupby(['Name'])['Name'].agg([len])
#video_games.sort_values(by = 'len', ascending=False)
#video_games

freq_abs = dataset['Platform'].value_counts().reset_index()
freq_abs.columns = ['Videogame', 'Frequência']

freq_rel = (dataset['Platform'].value_counts(normalize=True)*100
            ).reset_index()
freq_rel.columns = ['Videogame', 'Frequência Relativa (%)']

tabela_ditribuicao = pd.merge(freq_abs, freq_rel, on = 'Videogame')

tabela_ditribuicao = tabela_ditribuicao.sort_values('Frequência', ascending=False)

tabela_ditribuicao
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6));
plt.bar(tabela_ditribuicao['Videogame'], tabela_ditribuicao['Frequência'])
plt.xlabel("Videogame")
plt.ylabel("Frequência")
plt.title("Distribuição dos videogames")
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.barh(tabela_ditribuicao['Videogame'],tabela_ditribuicao['Frequência'])
plt.xlabel("Freqência")
plt.ylabel("Videogame")
plt.title("Distribuição dos videogames")
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.pie(tabela_ditribuicao['Frequência'], labels = tabela_ditribuicao['Videogame'], autopct="%1.1f%%")
plt.title("Distribuição dos videogames")
plt.axis("equal")
plt.tight_layout()

plt.show()
```

```python
dataset = pd.read_csv("/content/vgsales.csv")

dataset = dataset.loc[dataset['Platform'].isin(['PS2', 'PS3', 'X360'])]
dataset = dataset.loc[dataset['Genre'].isin(['Action', 'Adventure', 'Shooter', 'Sports'])]

#frq_absoluta = dataset.groupby(['Platform', 'Genre']).size().reset_index()
frq_absoluta = dataset.groupby(['Platform', 'Genre']).size().reset_index()
#frq_absoluta.columns = ['Videogame', 'Genêro', 'Frequência']

total = frq_absoluta['Frequência'].sum()
frq_absoluta['Frequência Relatica (%)'] = round(frq_absoluta['Frequência'] * 100 / total)

frq_absoluta
```

```python
dataset = pd.read_csv("/content/vgsales.csv")

filtered_data = dataset.loc[ dataset['Platform'].isin(['PS2', 'PS3', 'X360']) & dataset['Genre'].isin(['Action', 'Adventure', 'Shooter', 'Sports'])]

freq_table = pd.crosstab(index = filtered_data['Platform'], columns = filtered_data['Genre'])

labels = ['Ação do PS2', 'Aventura do PS2', 'Tiro do PS2', 'Esporte do PS2', 'Ação do PS3', 'Aventura do PS3', 'Tiro do PS3', 'Esporte do PS3',
          'Ação do X360', 'Aventura do X360', 'Tiro do X360', 'Esporte do X360']

plt.figure(figsize = (12,6))
plt.pie(freq_table.values.flatten(), autopct = '%1.1f%%', labels = labels, startangle = 90)
plt.title("Distribuição de jogos de videogame por gênero")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
