### Carregamento da base de dados

```python
from seaborn import load_dataset
import pandas as pd

df=load_dataset("titanic")

#df.info() #A coluna 'deck' tem apenas 203 valores non-null, assim eu vou pensar em tirar ela

df=df.drop('deck', axis=1) #exclusão da coluna 'deck'
df=df.dropna() # exclusão das linas com NaN

df=df.drop('class', axis=1) #exclusão da coluna 'class' que está diretamente relacionada com a coluna 'pclass'
df=df.drop('alive',axis=1) #exclusão da coluna 'alive' que está relacionada com a coluna 'survived'
#df.describe()
df
```

### OneHotEncoder

```python
#Tranformação das colunas categoricas
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

column_transformer = make_column_transformer((OneHotEncoder(),['sex','embarked','who','adult_male','embark_town','alone']), remainder='passthrough')

df=column_transformer.fit_transform(df)

df=pd.DataFrame(data=df, columns=column_transformer.get_feature_names_out())
```

```python
df
```

### Separação dos dados de teste e de treinamento

```python
y=df['remainder__pclass']

X=df.drop('remainder__pclass', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y) #80% treino e 20% teste
```

### Treinamento dos modelos

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

modelo_1=DecisionTreeClassifier(criterion='entropy', max_depth=5)
modelo_2=DecisionTreeClassifier(max_depth=4)

modelo_3=KNeighborsClassifier(n_neighbors=4, metric='euclidean', algorithm='brute')
modelo_4=KNeighborsClassifier(n_neighbors=7, metric='euclidean', algorithm='brute')

modelo_1.fit(X_treino, y_treino)
modelo_2.fit(X_treino, y_treino)
modelo_3.fit(X_treino, y_treino)
modelo_4.fit(X_treino, y_treino)
```

###Predições e Resultados

```python
from sklearn import metrics

resultado_1=modelo_1.predict(X_teste)
resultado_2=modelo_2.predict(X_teste)
resultado_3=modelo_3.predict(X_teste)
resultado_4=modelo_4.predict(X_teste)

porcentagens=[]

acc_1=metrics.accuracy_score(resultado_1,y_teste)
porcentagens.append(round(acc_1*100))

acc_2=metrics.accuracy_score(resultado_2,y_teste)
porcentagens.append(round(acc_2*100))

acc_3=metrics.accuracy_score(resultado_3,y_teste)
porcentagens.append(round(acc_3*100))

acc_4=metrics.accuracy_score(resultado_4,y_teste)
porcentagens.append(round(acc_4*100))

for i in range(4):
  print("{}%".format(porcentagens[i])+" modelo "+str(i+1))
```
