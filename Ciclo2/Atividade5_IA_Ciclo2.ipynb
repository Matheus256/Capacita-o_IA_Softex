```python
# 2) Carregar o dataset: "Car Crashes" do Seaborn. Dataset de Regressão

import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

data = sns.load_dataset("car_crashes")

# 3) Drop a coluna 'abbrev'
# 4) A coluna 'total' é a coluna target (objetivo)

y = data['total']

# 5) As features são as colunas de 1 a 6

X = data.iloc[ : , 1 : 6]
colunas = X.columns

# 6) Normalizar as seguintes colunas:

# 'speeding'
# 'alcohol'
# 'not_distracted'
# 'no_previous'
# 'ins_premium'

df = X.values
x_scale = MinMaxScaler().fit_transform(df)
X = pd.DataFrame(data=x_scale, columns=colunas)

X["ins_losses"] = data["ins_losses"]

display(X.head())

# 7) Separar os dados em treinamento (80%) e teste (20%)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None)
```

```python
# 8) Importar os pacotes do MLflow

import mlflow
import mlflow.sklearn

from sklearn.linear_model import LinearRegression
from sklearn import metrics

from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

# 9) Rodar 1 experimento com 3 execuções (RUNs)

exp = mlflow.set_experiment(experiment_name="Experimento_IA_Atividade_5")

# 10) Run 1:

# Instanciar um modelo Linear (sem parâmetros)
# Treinar
# Fazer o Predict
# Calcular o MSE, MAE e R2_SCORE
# Guardar as métricas: MSE, MAE e R2_SCORE
# Guardar o modelo no Tracking do MLflow (log_model)
# Finalizar a Execução

mlflow.end_run()

mlflow.start_run(run_name="Run_1", experiment_id= exp.experiment_id)

regressor1 = LinearRegression()
regressor1.fit(X_train, y_train)
resultados1 = regressor1.predict(X_test)

metricas = {}
metricas["mse"] = metrics.mean_squared_error(y_test, resultados1)
metricas["mae"] = metrics.mean_absolute_error(y_test, resultados1)
metricas["r2"] = metrics.r2_score(y_test, resultados1)

mlflow.log_metrics(metricas)

mlflow.sklearn.log_model(regressor1, "RegracaoLinear")

mlflow.end_run()

# 11) Run 2:

# Instanciar uma DecisionTreeRegressor
# Usar como parâmetros: criterion = 'friedman_mse' e max_depth = 4
# Treinar
# Fazer o Predict
# Calcular o MSE, MAE e R2_SCORE
# Guardar as métricas: MSE, MAE e R2_SCORE
# Guardar os parâmetros: Criterion e MAX_DEPTH
# Guardar o modelo no Tracking do MLflow (log_model)
# Finalizar a Execução

mlflow.start_run(run_name = "Run_2", experiment_id = exp.experiment_id)

regressor2 = DecisionTreeRegressor(criterion = 'friedman_mse', max_depth = 4)
regressor2.fit(X_train, y_train)
resultados2 = regressor2.predict(X_test)

metricas2 = {
    "mse" : metrics.mean_squared_error(y_test, resultados2),
    "mae" : metrics.mean_absolute_error(y_test, resultados2),
    "r2" : metrics.r2_score(y_test, resultados2)
}
mlflow.log_metrics(metricas2)

parametros2 = { "Criterion" : "friedman_mse", "MAX_DEPTH" : 4}
mlflow.log_params(parametros2)

mlflow.sklearn.log_model(regressor2, "DecisionTreeRegressor")

mlflow.end_run()

# 12) Run 3:

# Instanciar um KNeighborsRegressor
# Usar como parâmetro: n_neighbors=5
# Treinar
# Fazer o Predict
# Calcular o MSE, MAE e R2_SCORE
# Guardar as métricas: MSE, MAE e R2_SCORE
# Guardar os parâmetros: n_neighbors
# Guardar o modelo no Tracking do MLflow (log_model)
# Finalizar a Execução

mlflow.start_run(run_name = "Run_3", experiment_id = exp.experiment_id)

regressor3 = KNeighborsRegressor(n_neighbors = 5)
regressor3.fit(X_train, y_train)
resultados3 = regressor3.predict(X_test)

metricas3 = {
    "mse" : metrics.mean_squared_error(y_test, resultados3),
    "mae" : metrics.mean_absolute_error(y_test, resultados3),
    "r2" : metrics.r2_score(y_test, resultados3)
}
mlflow.log_metrics(metricas3)

mlflow.log_param("N_Neighbors", 5)

mlflow.sklearn.log_model(regressor3, "KNeighborsRegressor")

mlflow.end_run()
```
